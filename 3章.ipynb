```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def step_function(x):
    return np.array(x>0, dtype=np.int)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
def softmax(a):
    a_max = np.max(a)
    exp_a = np.exp(a - a_max) #オーバーフロー対策
    sum_exp_a = np.sum(exp_a)
    
    return exp_a / sum_exp_a
```
